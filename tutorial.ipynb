```python
import matplotlib.pyplot as plt
import numpy as np
scores=np.random.normal(loc=70, scale=10, size=200)
plt.hist(scores,bins=20,edgecolor="black")
plt.axvline(scores.mean(),linestyle="--",color="red")
plt.tight_layout()
plt.xlabel("scores")
plt.ylabel("frequency")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
penguin=sns.load_dataset("penguins")
penguin.head()
sns.scatterplot(
    data=penguin,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species"    
)
plt.title("penguins dataset")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips=sns.load_dataset("tips")
tips.head()
sns.boxplot(
    data=tips,
    x="total_bill",
    y="day",
    hue="sex"
)
plt.title("tips data")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
tips.head(10)
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
   style="smoker"
)
plt.title("Tips data (total_bills vs tip)")
plt.xlabel("total_bill")
plt.ylabel("tips")
plt.grid()
plt.show()
```
